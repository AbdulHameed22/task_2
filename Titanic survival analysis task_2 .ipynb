```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_df = pd.read_csv("C:/Users/NHI652/Documents/gender_submission.csv")
titanic_df
pd.set_option('display.max_rows',1000)
# Replacing numeric values in the 'Survived' column with labels. '1'- Survived; '0'- Not Survived
titanic_df['Survived'] = titanic_df['Survived'].replace({0: 'Not Survived', 1: 'Survived'})
titanic_df
```

```python
# Dataset Preview and Stats
print(titanic_df.describe())
```

```python
# Number of occurrences of each unique value in 'Survived'
survived_counts = titanic_df['Survived'].value_counts()

# Pie chart with labels inside
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(survived_counts, labels=survived_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightgreen'])

# Drawing a circle at the center of the pie chart to make it look like a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Display total distribution and count of passengers who survived
total_passengers = len(titanic_df)
survived_passengers = survived_counts[1] 

plt.text(0, 0, f'Total Passengers: {total_passengers}\nSurvived: {survived_passengers}', ha='center', va='center', fontsize=12)

plt.title('Distribution of Survived')
plt.show()
```

```python
# Plotting the distribution of survived vs. not survived
survival_counts = titanic_df['Survived'].value_counts()
plt.figure(figsize=(8, 6))
survival_counts.plot(kind='bar', color=['lightcoral', 'lightgreen'])
plt.xlabel('Survival Status')
plt.ylabel('Count')
plt.title('Distribution of Survived vs. Not Survived')
plt.show()
```

```python
import numpy as np
# Assuming 'titanic_df' is given DataFrame containing 'PassengerID' and 'Survived' columns
# Replace this with your actual DataFrame if needed
titanic_df = pd.read_csv("C:/Users/NHI652/Documents/gender_submission.csv")
titanic_df
pd.set_option('display.max_rows',1000)

# Setting a seed for reproducibility
np.random.seed(42)

# Sample size
sample_size = 100  # Adjust the sample size as needed

# Take a random sample of the dataset
random_sample = titanic_df.sample(n=sample_size)

# Calculating the probability of survival
survival_probability = random_sample['Survived'].mean()

# Calculating the probability of not surviving (death)
death_probability = 1 - survival_probability

# Printing the results
print(f'Probability of Survival: {survival_probability:.2%}')
print(f'Probability of Death: {death_probability:.2%}')
```
